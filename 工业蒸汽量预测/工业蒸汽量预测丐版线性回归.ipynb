```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
with open('./zhengqi_train.txt') as fr:
    train_data = pd.read_table(fr,sep="\t")
with open('./zhengqi_test.txt') as fr_text:
    test_data = pd.read_table(fr_text,sep="\t")
```

```python
plt.figure(figsize = (20,16))
colnm = train_data.columns.tolist()
mcorr = train_data[colnm].corr(method="spearman")
mask = np.zeros_like(mcorr,dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
cmap = sns.diverging_palette(220,10,as_cmap=True)
g = sns.heatmap(mcorr,mask = mask,cmap = cmap,square = True, annot = True, fmt = '0.2f')
plt.show()
```

```python
name_drop = ['V5','V6','V7','V9','V10','V11','V13','V14','V15','V17','V18',
             'V19','V20','V21','V22','V23','V24','V25','V26','V28','V29',
            'V30','V32','V33','V34','V35','V36','target']
name_extract = [col for col in train_data.columns if col not in name_drop]
```

```python
X_data = train_data[name_extract]
Y_data = train_data['target']
X_test = test_data[name_extract]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
linreg.fit(X_data,Y_data)
```

```python
linreg.score(X_data,Y_data)
```

```python
realy = linreg.predict(X_test)
realy
```

```python
realy = pd.DataFrame(realy)
```
