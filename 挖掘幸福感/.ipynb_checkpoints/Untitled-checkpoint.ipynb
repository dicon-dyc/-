```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder,StandardScaler,MinMaxScaler
import lightgbm as lgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import svm
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import KFold
import collections
import datetime
```

```python
train_data = pd.read_csv('happiness_train_complete.csv',parse_dates = ["survey_time"],encoding = 'latin-1')
test_data = pd.read_csv('happiness_test_complete.csv',parse_dates=["survey_time"],encoding = 'latin-1')
```

```python
train_data.head(10)
```

**<font size = 6 color = "#414141">1.数据预处理</font>**

**<font size = 4 color = "#414141">1.1删除缺失率大于0.25的特征</font>**

```python
drop_features = []
for col in train_data.columns:
    if train_data[col].isnull().sum()/len(train_data[col]) > 0.25:
        drop_features.append(col)
```

```python
extract_features = [col for col in train_data.columns if col not in drop_features]
train_data1 = train_data[extract_features]
extract_features.remove('happiness')
test_data1 = test_data[extract_features]
```

```python
for col in train_data1.columns:
    train_data1[col]=train_data1[col].map(lambda x:3 if x==-8 else x)
for col in test_data1.columns:
    test_data1[col]=test_data1[col].map(lambda x:3 if x== -8 else x)
```

**<font size = 4 color = "#414141">1.2根据众数填补缺失值</font>**

```python
for col in train_data1.columns:
    if train_data1[col].isnull().sum() != 0:
        meb_counts = collections.Counter(train_data1[col].values)
        most_meb = meb_counts.most_common(1)[0][0]
        train_data1[col].fillna(most_meb,inplace = True)
for col in test_data1.columns:
    if test_data1[col].isnull().sum() != 0:
        meb_counts = collections.Counter(test_data1[col].values)
        most_meb = meb_counts.most_common(1)[0][0]
        test_data1[col].fillna(most_meb,inplace = True)
```

**<font size = 6 color = "#414141">2.特征工程</font>**

**<font size = 4 color = "#414141">2.1查看测试集和训练集数据分布的差异,并且删除分布差异大的特征</font>**

```python
for col in test_data1.columns:
    g = sns.kdeplot(train_data1[col],color = "Red",shade = True)
    g = sns.kdeplot(test_data1[col],color = "Blue",shade = True)
    g.set_xlabel(col)
    g.set_ylabel('frequency')
    g = g.legend(["train","test"])
    plt.show()
```

**<font size = 4 color = "#414141">2.2删除相关性低于0.05的特征</font>**

```python
train_data1.corr()['happiness'][abs(train_data.corr()['happiness'])<0.05]
```

```python
features = train_data1.corr()['happiness'][abs(train_data.corr()['happiness'])<0.05]
features = features.index.tolist()
extract_features = [col for col in train_data1.columns if col not in features]
tmp = ['birth','s_birth','f_birth','m_birth']
for i in tmp:
    if i not in extract_features:
        extract_features.append(i)
train_data1 = train_data1[extract_features]
extract_features.remove('happiness')
test_data1 = test_data1[extract_features]
```

**<font size = 4 color = "#414141">2.3将所有的出生日期特征构造成年龄特征</font>**

```python
for i in train_data1['survey_time'].index:
    train_data1['survey_time'][i] = pd.to_datetime(train_data1['survey_time'][i]).year
for i in test_data1['survey_time'].index:
    test_data1['survey_time'][i] = pd.to_datetime(test_data1['survey_time'][i]).year
```

```python
train_data1.columns
```

```python
for i in train_data1['birth'].index:
    train_data1['birth'][i] = train_data1['survey_time'][i]-train_data1['birth'][i]
    train_data1['s_birth'][i] = train_data1['survey_time'][i]-train_data1['s_birth'][i]
    train_data1['f_birth'][i] = train_data1['survey_time'][i]-train_data1['f_birth'][i]  
    train_data1['m_birth'][i] = train_data1['survey_time'][i]-train_data1['m_birth'][i] 
for i in test_data1['birth'].index:
    test_data1['birth'][i] = test_data1['survey_time'][i]-test_data1['birth'][i]
    test_data1['s_birth'][i] = test_data1['survey_time'][i]-test_data1['s_birth'][i]
    test_data1['f_birth'][i] = test_data1['survey_time'][i]-test_data1['f_birth'][i]  
    test_data1['m_birth'][i] = test_data1['survey_time'][i]-test_data1['m_birth'][i]   
```

```python
train_data1 = train_data1.rename(columns = {'birth':'age','s_birth':'s_age','f_birth':'f_age','m_birth':'m_age'})
test_data1 = test_data1.rename(columns = {'birth':'age','s_birth':'s_age','f_birth':'f_age','m_birth':'m_age'})
```

**<font size = 4 color = "#414141">2.4将离散个数小于等于10个不应具有大小关系的特征列取出准备进行onehot编码</font>**

```python
classifier_features = []
for col in train_data1.columns:
    if len(train_data1[col].unique())<= 10:
           classifier_features.append(col)
print(classifier_features)
#for col in classifier_features:
 #          train_data1.drop(columns = col)
  #         test_data1.drop(columns = col)
```

```python
classifier_features = [col for col in classifier_features if col not in ['happiness','id','city','county']]
```

```python
onehot_data = pd.concat((train_data1,test_data1),axis = 0,join = 'outer')
onehot_data = onehot_data[classifier_features]
```

**<font size = 4 color = "#414141">2.5进行onehot编码</font>**

```python
onehotencoder = OneHotEncoder()
onehot_data = onehotencoder.fit_transform(onehot_data).toarray()
```
