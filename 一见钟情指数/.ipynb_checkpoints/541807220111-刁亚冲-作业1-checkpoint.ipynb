```python
def computeCost(X,y,theta):
    inner = np.power(((X*theta.T)-y),2)
    return np.sum(inner)/(2*len(X))
```

```python
X_data.insert(0,'Ones',1)
X_data.shape
```

```python
X_data.head()
```

```python
X_data = np.matrix(X_data.values)
Y_data = np.matrix(Y_data.values)
```

```python
theta = np.matrix(np.zeros((1,19)))
```

```python
theta.shape
```

```python
def gradientDescent(X_data,Y_data,theta,alpha,iters):
    temp = np.matrix(np.zeros(theta.shape))
    parameters = int(theta.ravel().shape[1])
    cost = np.zeros(iters)
    
    for i in range(iters):
        error  = (X_data * theta.T) - Y_data
        
        for j in range(parameters):
            term = np.multiply(error,X_data[:,j])
            temp[0,j] = theta[0,j] - ((alpha/len(X_data))*np.sum(term))
        
        theta = temp
        cost[i] = computeCost(X_data,Y_data,theta)
    return theta, cost
```

```python
alpha = 0.06
iters = 1000
```

```python
g,cost = gradientDescent(X_data,Y_data.T,theta,alpha,iters)
g
```

```python
computeCost(X_data,Y_data.T,g)
```

```python
fig,ax = plt.subplots(figsize=(12,8))
ax.plot(np.arange(iters),cost,'r')
ax.set_xlabel('Iterations')
ax.set_ylabel('Error vs. Training Epoch')
plt.show()
```
